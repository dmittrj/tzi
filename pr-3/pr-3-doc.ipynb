```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 12*np.pi, 0.1)
plt.figure(figsize=(10, 2))
plt.plot(np.arange(0, 12*np.pi, 0.1), np.sin(x))

plt.xlabel('t')
plt.ylabel('cos(t)')
plt.grid(True, color='lightgray')

plt.show()
```

```python
t = np.linspace(0, 1, 1000)

carrier_signal = np.cos(2 * np.pi * 40 * t)
modulating_signal = np.cos(2 * np.pi * 2 * t)
output_signal = (1 + 0.5 * modulating_signal) * carrier_signal

plt.figure(figsize=(10, 6))
plt.subplot(3, 1, 1)
plt.plot(t, carrier_signal)
plt.title('Несущий сигнал')

plt.subplot(3, 1, 2)
plt.plot(t, modulating_signal)
plt.title('Модулирующий сигнал')

plt.subplot(3, 1, 3)
plt.plot(t, output_signal)
plt.title('Сигнал на выходе (амплитудно-модулированный сигнал)')

plt.tight_layout()
plt.show()
```

```python
t = np.linspace(0, 1, 1000)

output_signal1 = (1 + 0.25 * modulating_signal) * carrier_signal
output_signal2 = (1 + 0.75 * modulating_signal) * carrier_signal
output_signal3 = (1 + 1.25 * modulating_signal) * carrier_signal

plt.figure(figsize=(10, 6))
plt.subplot(3, 1, 1)
plt.plot(t, output_signal1)
plt.title('Модуляция 25% (m = 0.25)')

plt.subplot(3, 1, 2)
plt.plot(t, output_signal2)
plt.title('Модуляция 75% (m = 0.75)')

plt.subplot(3, 1, 3)
plt.plot(t, output_signal3)
plt.title('Модуляция 125% (m = 1.25)')

plt.tight_layout()
plt.show()
```

```python
frequency_carrier = 200  # Частота несущего сигнала
frequency_modulating = 50  # Частота модулирующего сигнала
amplitude_carrier = 20  # Амплитуда несущего сигнала
amplitude_modulating = 10  # Амплитуда модулирующего сигнала
modulation_index = 0.7  # Коэффициент модуляции

# Временная ось
time = np.linspace(0, 1, 10000)

# Генерация несущего сигнала
carrier = amplitude_carrier * np.cos(2 * np.pi * frequency_carrier * time)

# Генерация модулирующего сигнала
modulating = amplitude_modulating * np.cos(2 * np.pi * frequency_modulating * time)

# Амплитудно-модулированный сигнал
modulated = (1 + modulation_index / amplitude_modulating * modulating) * carrier

# Спектр амплитудно-модулированного сигнала
spectrum = np.fft.fft(modulated) / len(modulated) * 2

# Построение графиков
freq = np.fft.fftfreq(len(modulated), 1 / len(modulated))
plt.plot(freq, np.abs(spectrum))
plt.xlim(0, len(modulated) / 2.5 / len(time) * 1000)
plt.xlabel('Частота')
plt.ylabel('Магнитуда (спектральная мощность)')
plt.title('Спектр амплитудно-модулированного сигнала')

plt.tight_layout()
plt.grid(True, color='lightgray')
plt.show()
```

```python
frequency_carrier = 100  # Частота несущего сигнала
frequency_modulating = 10  # Частота модулирующего сигнала
amplitude_carrier = 15  # Амплитуда несущего сигнала
amplitude_modulating = 10  # Амплитуда модулирующего сигнала
modulation_index = 0  # Коэффициент модуляции

# Временная ось
time = np.linspace(0, 1, 100000)

# Генерация несущего сигнала
carrier = amplitude_carrier * np.cos(2 * np.pi * frequency_carrier * time)

# Генерация модулирующего сигнала
modulating = amplitude_modulating * np.cos(2 * np.pi * frequency_modulating * time)

# Амплитудно-модулированный сигнал
#modulated = (amplitude_carrier + modulation_index * modulating) * np.cos(2 * np.pi * frequency_carrier * time)
modulated = (1 + modulation_index / amplitude_modulating * modulating) * carrier

# Спектр амплитудно-модулированного сигнала
spectrum = np.fft.fft(modulated) / len(modulated) * 2

# Построение графиков
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(time, modulated)
plt.xlabel('Время')
plt.ylabel('Амплитуда')
plt.title('Амплитудно-модулированный сигнал')

plt.subplot(1, 2, 2)
freq = np.fft.fftfreq(len(modulated), 1 / len(modulated))
plt.plot(freq, np.abs(spectrum))
plt.xlim(0, len(modulated) / 5 / len(time) * 1000)
plt.xlabel('Частота')
plt.ylabel('Магнитуда')
plt.title('Спектр')

plt.tight_layout()
plt.show()
```
